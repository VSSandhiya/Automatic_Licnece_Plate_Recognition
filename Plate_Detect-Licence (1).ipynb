```python
from ultralytics import YOLO
import cv2, numpy as np, re
from collections import Counter

m=YOLO("/content/plate_yolov8.pt")  # change path if needed
```

```python
import easyocr
r=easyocr.Reader(['en'],gpu=True)
```

```python
def clean_plate(s):
    s=s.upper()
    s=re.sub(r'[^A-Z0-9]','',s)
    return s

def looks_like_plate(s):
    # Basic Indian plate pattern: XX00XX0000 or similar
    # Accept 8 to 11 chars to be flexible
    if len(s)<8 or len(s)>11: return False
    if not re.match(r'^[A-Z]{2}\d{1,2}[A-Z]{1,3}\d{3,4}$',s):
        return False
    return True
```

```python
def variants(img):
    v=[]
    g=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    v.append(img)                           # original
    v.append(cv2.resize(img,None,fx=2,fy=2,interpolation=cv2.INTER_CUBIC))  # upscale

    b=cv2.GaussianBlur(g,(3,3),0)
    v.append(cv2.cvtColor(b,cv2.COLOR_GRAY2BGR))

    th=cv2.adaptiveThreshold(g,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                             cv2.THRESH_BINARY,11,2)
    v.append(cv2.cvtColor(th,cv2.COLOR_GRAY2BGR))

    k=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
    sh=cv2.filter2D(img,-1,k)
    v.append(sh)

    return v

def ocr_vote(plate_img):
    cand=[]
    for im in variants(plate_img):
        out=r.readtext(im,detail=1)
        for bb,txt,conf in out:
            t=clean_plate(txt)
            if len(t)>=6:  # ignore tiny junk
                cand.append((t,conf))

    if not cand:
        return "",0.0

    # Prefer valid Indian-like plates first
    valid=[(t,c) for t,c in cand if looks_like_plate(t)]
    use=valid if valid else cand

    # Weighted vote (sum confidence per string)
    score={}
    for t,c in use:
        score[t]=score.get(t,0)+c

    best=max(score,key=score.get)
    return best,score[best]/max(1,len([x for x in use if x[0]==best]))
```

```python
import glob, cv2
from google.colab.patches import cv2_imshow

paths=sorted(glob.glob("/content/Indian-License-Plate-1/test/images/*"))
img_path=paths[0]
print("Using:",img_path)

im=cv2.imread(img_path)
if im is None: raise Exception("Image not loaded")

res=m.predict(source=im,conf=0.25,iou=0.5,verbose=False)[0]

best_plate=""
best_conf=0
best_box=None

for b in res.boxes:
    x1,y1,x2,y2=map(int,b.xyxy[0].tolist())
    crop=im[max(0,y1):max(0,y2),max(0,x1):max(0,x2)]
    if crop.size==0:
        continue
    txt,txt_conf=ocr_vote(crop)
    if txt_conf>best_conf:
        best_conf=txt_conf
        best_plate=txt
        best_box=(x1,y1,x2,y2)

out=im.copy()
if best_box:
    x1,y1,x2,y2=best_box
    cv2.rectangle(out,(x1,y1),(x2,y2),(0,255,0),2)
    cv2.putText(out,f"{best_plate} ({best_conf:.2f})",(x1,max(30,y1-10)),
                cv2.FONT_HERSHEY_SIMPLEX,0.9,(0,255,0),2)

print("Final Plate:",best_plate)
print("OCR Confidence:",best_conf)
cv2_imshow(out)
```
